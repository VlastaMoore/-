```python
#Задача 4.2 
import pandas as pd
import numpy as np
import random

#Количество попыток
n=10000

#Бросок первого кубика
cube1 = np.random.randint(6, size=n)
print((f'"cube1="{cube1}'))

#Бросок второго кубика
cube2 = np.random.randint(6, size=n)
print((f'"cube2="{cube2}'))

df = pd.DataFrame(list(zip(cube1, cube2)), columns=["cube1", "cube2"])
variations = df.groupby(df.columns.tolist(),as_index=False).size() # множество элементарных событий и кол-во их повторений

#Интересующие события

df["sum"] = df['cube1'] + df['cube2'] >= 9

B = df['sum'].value_counts()[True]/n
print('Вероятность суммы выпавших очков не меньше 11 равна', B)

print(variations)
```

```python
#Задача 9

#Количество карт
NUMBER_CARDS = 36
#Количество мастей
NUMBER_SUIT = 4

#Событие А - выпадет 6, а потом 7
probability = (NUMBER_SUIT/NUMBER_CARDS) * (NUMBER_SUIT/(NUMBER_CARDS-1))
print("Вероятность того, что 1-й картой была 6-ка, а 2-й — 7-ка: ", end="")
print(format(probability, '.4f'))
print()

#Событие B - выпадут две бубны
diamond = NUMBER_CARDS/NUMBER_SUIT  #Количество бубны в колоде
probability = (diamond/NUMBER_CARDS) * ((diamond-1)/(NUMBER_CARDS-1))
print("Вероятность того, что обе карты бубновой масти: ", end="")
print(format(probability, '.4f'))
print()

#Событие AB - выпадет бубновая 6, а потом бубновая 7
probability_AB = (1/NUMBER_CARDS)*(1/(NUMBER_CARDS-1))

#Событие A|B - выпадет 6, а потом 7 при условии, что обе бубновой масти
probability_A_if_B = probability_AB / probability
print("Вероятность того, что 1-й картой была 6-ка, а 2-й — 7-ка ")
print("при условии, что обе карты бубновой масти: ", end="")
print(format(probability_A_if_B, '.4f'))
```
